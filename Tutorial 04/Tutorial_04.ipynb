Part 1

 Your first task is to make two plots:

(a) strain vs time  
(b) frequency vs time

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
GW_data = strain, freq, time = np.loadtxt('GW_data_file.csv', delimiter=',')
#GW_data = np.loadtxt('GW_data_file.csv', delimiter=',')
np.set_printoptions(precision=5, suppress=True, floatmode='fixed')
#print(strain, freq, time)
print(strain, freq, time)
```

---------------------------------------------------------------------------------------------------------------------------------------------

a) Plot of strain v/s time

```python
plt.figure(figsize=(12,8))
plt.plot(time, strain,color='green')
plt.grid(True, linestyle='-', color='gray')
plt.xlabel('Time(s)')
plt.ylabel('Strain')
plt.title('Strain v/s Time')
```

-----------------------------------------------------------------------------------------------------------------------------------------------

b) Plot of frequency v/s time

```python
plt.figure(figsize=(12,8))
plt.plot(time, freq, color = 'black')
plt.grid(True, linestyle='-', color='gray')
plt.xlabel('Time(s)')
plt.ylabel('Frequency(Hz)')
plt.title('Frequency v/s Time')
```

Part 2

You should notice something weird about the frequency plot. Why is it weird? Find out about it under the next section.

The frequency plot has these weird step-like features. This is due to the fact that the data is divided into segments because the frequency varies with time. The frequency component in the given data has the highest amplitude for a given segment. Therefore, steps are seen in the plot of frequency v/s time.

Part 3

Find out the indices of zero-crossings. Use the frequency and the time at these zero-crossings to make a frequency vs time scatter plot.

```python
zero_crossing_indices = []

#print(len(strain))

for i in range(1, len(strain)):
    if (strain[i-1] < 0 and strain[i] >= 0) or (strain[i-1] > 0 and strain[i] <= 0):
        zero_crossing_indices.append(i)


#print(len(zero_crossing_indices))

print(f"The indices of the zero-crossings are {zero_crossing_indices}.")
```

```python
strain_zero_crossing_values = strain[zero_crossing_indices]
print(strain_zero_crossing_values)
freq_zero_crossing_values = freq[zero_crossing_indices]
print(freq_zero_crossing_values)
time_zero_crossing_values = time[zero_crossing_indices]
print(time_zero_crossing_values)
```

```python
plt.figure(figsize=(10,6))
plt.grid(True)
plt.scatter(time_zero_crossing_values, freq_zero_crossing_values, s = 40, color = 'green')

plt.xlabel("Time(s)")
plt.ylabel("Frequency(Hz)")
plt.title("Plot of Frequency v/s Time at zero-crossings")
```
