Part 1

Exclude all the stars with low probability of belonging to the cluster, and then calculate distance for each star, and find the mean.

```python
Beehive_data = np.loadtxt("Beehive_data.csv", delimiter = ",")
print(Beehive_data)
print(len(Beehive_data))
```

```python
Beehive_data_rounded = np.round(Beehive_data,1)
print(Beehive_data_rounded)
print(len(Beehive_data_rounded))
```

```python
low_prob = np.where(Beehive_data_rounded[:,2]<50)
Beehive_data_updated = np.delete(Beehive_data_rounded,low_prob,axis=0)
print(Beehive_data_updated)
print(len(Beehive_data_updated))
```

```python
def distance_to_Beehive(Beehive_data_updated):
    power_raise_to_10 = ((Beehive_data_updated[:,0] + 2.5*(Beehive_data_updated[:,1]) + 0.17)/5)
    d = 10**power_raise_to_10
    return d
    
result = distance_to_Beehive(Beehive_data_updated)
print(result)
print(len(result))
```

```python
result_rounded = np.round(result,3)
print(result_rounded)
print(len(result_rounded))
```

```python
distance_to_Beehive_cluster = result.mean()
print(distance_to_Beehive_cluster)
```

```python
distance_to_Beehive_cluster = result_rounded.mean()
print(np.round(distance_to_Beehive_cluster,3))
```

```python
sum_of_dist = 0
for d in result_rounded:
    sum_of_dist = sum_of_dist + d
distance_to_Beehive = sum_of_dist/result_rounded.size
print(distance_to_Beehive)
print(np.round(distance_to_Beehive,3))
```
